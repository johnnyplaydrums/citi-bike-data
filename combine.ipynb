### Concat and Glob
We can read all the given files in our set path as shown below and read them into our df called summer 2015.
from there we run a for loop that will take all possible files in the path and append them together. Finally we use pandas built in concat to combine them together.

```python
path =r'C:\Users\Sammauel\Documents\DataScience\Citibike' # use your path
allFiles = glob.glob(path + "/*.csv")
summer2015 = pd.DataFrame()
list_ = []
for file_ in allFiles:
    df = pd.read_csv(file_,index_col=None, header=0)
    list_.append(df)
summer2015 = pd.concat(list_)
```

```python
summer2015.drop('Unnamed: 0', axis=1, inplace=True)
summer2015.tail()
```
